```python
import os
import pandas as pd

# cloud storage folder
CLOUD_FOLDER = "/content/drive/MyDrive/Cloud_Heart_Project/"
RAW_DATA = CLOUD_FOLDER + "heart.csv"
PROCESSED_DATA = CLOUD_FOLDER + "processed_high_risk_patients.csv"
BACKUP_FOLDER = CLOUD_FOLDER + "backups/"

#backup folder
os.makedirs(BACKUP_FOLDER, exist_ok=True)
```

```python
# Load Data from Cloud storage
print("\n📂 Loading raw data from cloud storage...")
df = pd.read_csv(RAW_DATA)
print(f"✅ Raw data loaded:{df.shape[0]} rows, {df.shape[1]} columns")
```

```python
# View Files in Cloud storage
print("\n📂 Files currwntly in your cloud storage:")
for file in os.listdir(CLOUD_FOLDER):
    size = os.path.getsize(CLOUD_FOLDER + file) / 1024
    print(f" - {file} ({round(size, 2)} KB)")
```

```python
# Process data(SF version__No error)
print("\n⚙️ Processing data in cloud...")
risk_column = df.columns[-1]
high_risk = df[df[risk_column] == 1]

print(f"✅ High-risk patients found: {len(high_risk)}")
```

```python
# Saved processed data back to Cloud
high_risk.to_csv(PROCESSED_DATA, index=False)
print(f"\n💾 Processed data saved to cloud: {PROCESSED_DATA}")
```

```python
# Backup In Cloud Storage
backup_file = BACKUP_FOLDER + "raw_data_backup.csv"
df.to_csv(backup_file, index=False)
print(f"✅ Backup saved to cloud: {backup_file}")
```

```python
print("\n" + "="*60)
print("📊 CLOUD STORAGE SUMMARY - PROJECT READY")
print("="*60)
print(f"Cloud Storage Location: {CLOUD_FOLDER}")
print(f"Raw Dataset in Cloud: {os.path.basename(RAW_DATA)}")
print(f"Processed File in Cloud: {os.path.basename(PROCESSED_DATA)}")
print(f"Backup Folder in Cloud: backups/")
print(f"Total files in storage: {len(os.listdir(CLOUD_FOLDER))}")
print("\n ALL CLOUD STORAGE TASKS COMPLETED!")
print("="*60)
```
